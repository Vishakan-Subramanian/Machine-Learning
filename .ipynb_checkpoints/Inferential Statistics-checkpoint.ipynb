```python
#Hypothesis Testing
cars_data = pd.read_csv("inferential_cars.csv")
cars = cars_data.copy()
```

```python
cars = cars[(cars.yearOfRegistration <= 2018)
           & (cars.yearOfRegistration >= 1950)
           & (cars.price >= 100)
           & (cars.price <= 150000)
           & (cars.powerPS >= 10)
           & (cars.powerPS <= 500)]
```

```python
#one sample test for mean
#Null Hyp : Avg price of car is 6000
#Alt Hyp : It's not
sample_size = 1000
sample1 = cars.sample(sample_size,random_state=0)
#random_state = 0 so that the same set of samples are given each time
pos_mean = 6000
print(sample1.price.mean())
```

```python
from scipy.stats import ttest_1samp
#t test for one sample
statistic, pvalue = ttest_1samp(sample1.price,pos_mean)
print(statistic,pvalue)
```

```python
#compute critical value
alpha = 0.05

#degrees of freedom (no. of indep. obs. in data)
n = len(cars.price)
df = n-1
print(n,df)
```

```python
from scipy.stats import t
#import t distribution
#point percentile function
cv = t.ppf([alpha/2,1-alpha/2], df)
#critical val
print(cv)
```

```python
#decision making:
# test values lie within crit values. Dont reject Null Hyp
# alpha < p, therefore, DO NOT REJECT NULL HYP.
# conclude mean = 6000.
```

```python
#one sample test for proportion
# Null Hyp : % of used car with automatic gear is 23
# Alt. Hyp : Not True
from statsmodels.stats.proportion import proportions_ztest
#total automatic
count = sample1.gearbox.value_counts()[1]
#total obs
nobs = len(sample1.gearbox)
#hyp value
p0 = 0.23
```

```python
#in the sample
sample1.gearbox.value_counts()/nobs
```

```python
statistic_oneprop, pvalue_oneprop = proportions_ztest(count = count, nobs = nobs, value = p0, alternative =  'two-sided', prop_var = False)
print(statistic_oneprop, pvalue_oneprop)
```

```python
#import normal dist
from scipy.stats import norm
#no need of degree of freedom since normal dist
#crit values
cv_norm  = norm.ppf([alpha/2,1-alpha/2]) #limits of PPF
print(cv_norm)
```

```python
# p = 0.22 > alpha
# z value is in crit val interval 
# therefore do not reject NULL hypothesis
# conclude % is 23.
```

```python
#two sample test for means
#is mean price of cars 30k-60k same as 70k-90k km?
#Null Hyp  : Equal
#Alt Hyp   : Not Equal

#first we need to check variance
km_7090 = cars[(cars.kilometer <= 90000) & (cars.kilometer >= 70000)]
km_3060 = cars[(cars.kilometer <= 60000) & (cars.kilometer >= 30000)]
sample_7090 = km_7090.sample(500, random_state = 0)
sample_3060 = km_3060.sample(500, random_state = 0)
```

```python
print(sample_7090.price.var(),sample_3060.price.var(),sample_7090.price.mean(),sample_3060.price.mean(),sep = "\n")
```

```python
# f statistic
from scipy.stats import f
F = sample_7090.price.var()/sample_3060.price.var()
print(F)
#therefore variances are unequal(Ratio isnt 1)
```

```python
df2 = len(sample_7090)-1
df1 = len(sample_3060)-1

print(scipy.stats.f.cdf(F,df1,df2))
#p value
f.ppf([alpha/2,1-alpha/2],df1,df2)
#crit values
```

```python
#variances unequal
#f stat 0.56 outside crit interval
#p <<< alpha
#Reject Null Hypothesis for variances
#Therefore do Welch t Test for unequal variances
```

```python
from scipy.stats import ttest_ind
statistic_twomean, pvalue_twomean = ttest_ind(sample_3060.price, sample_7090.price, equal_var = False)
print(statistic_twomean, pvalue_twomean, sep="\n")
```

```python
n1 = len(sample_3060)
n2 = len(sample_7090)
s12 = sample_3060.price.var()
s22 = sample_7090.price.var()
df = (((s12/n1)+(s22/n2))**2)/((((s12/n1)**2)/(n1-1))+(((s22/n2)**2)/(n2-1)))
df
```

```python
cv_t = t.ppf([alpha/2, 1-alpha/2],df)
print(cv_t)
#when df is large, cv_norm =~ cv_t
```

```python
#conclude Rejection of Null Hyp
# as p <<<< alpha
# and t statistic doesn't lie inside the crit value range
# both prices aren't the same
```

```python
#two sample test proportion
#are proportion of petrol cars in 2009-2013 and 2014-2018 diff?
# pi1 = pi2 null hyp(proportions equal)
# not equal alt hyp
```

```python
year_1418 = cars[(cars.yearOfRegistration <=2018)&(cars.yearOfRegistration >=2014)]
year_0913 = cars[(cars.yearOfRegistration <=2013)&(cars.yearOfRegistration >=2009)]
sample_1418 = year_1418.sample(1000,random_state=0)
sample_0913 = year_0913.sample(1000,random_state=0)
```

```python
from statsmodels.stats.proportion import proportions_ztest
count = [(sample_1418.fuelType == 'petrol').sum(), (sample_0913.fuelType == 'petrol').sum()]
nobs = [len(sample_1418), len(sample_0913)]
print(count[0]/nobs[0], count[1]/nobs[1])
#p hat values
```

```python
statistic, pvalue = proportions_ztest(count = count, nobs = nobs, value =0, alternative = 'two-sided', prop_var = False)
print(statistic,pvalue)
```
